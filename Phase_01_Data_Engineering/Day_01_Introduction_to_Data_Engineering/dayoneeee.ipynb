```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
top_student = df.loc[df['total_score'].idxmax()]
least_student = df.loc[df['total_score'].idxmin()]
print("Top student:")
print(top_student[['name','total_score']])
print("\nLeast student:")
print(least_student[['name','total_score']])
```

```python

total_students = df.shape[0]

total_columns = df.shape[1]

unique_departments = df['department'].nunique()

print(f"Total number of students: {total_students}")
print(f"Total number of columns: {total_columns}")
print(f"Number of unique departments: {unique_departments}")
```

```python
department_counts = df['department'].value_counts()
print("Count of students in each department:")
print(department_counts)
```

```python

highest_math_score = df['math_score'].max()


lowest_math_score = df['math_score'].min()


avg_math_score = df['math_score'].mean()


avg_programming_score = df['programming_score'].mean()


avg_attendance_percentage = df['attendance_percentage'].mean()

print(f"Highest Math Score: {highest_math_score}")
print(f"Lowest Math Score: {lowest_math_score}")
print(f"Average Math Score: {avg_math_score:.2f}")
print(f"Average Programming Score: {avg_programming_score:.2f}")
print(f"Average Attendance Percentage: {avg_attendance_percentage:.2f}%")
```

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
print("Total score calculated and added to DataFrame.")
```

```python
avg_total_score_by_gender = df.groupby('gender')['total_score'].mean()
print("Average Total Score by Gender:")
print(avg_total_score_by_gender)
```

```python
top_5_students = df.sort_values(by='total_score', ascending=False).head(5)
print("Top 5 students based on total score:")
print(top_5_students[['name', 'department', 'total_score']])
```

```python
low_attendance_students = df[df['attendance_percentage'] < 75]

if not low_attendance_students.empty:
    print("Students with attendance below 75%:")
    print(low_attendance_students[['name', 'department', 'attendance_percentage']])
else:
    print("All students have attendance above 75%.")
```

```python
electronics_students_count = df[df['department'] == 'Electronics'].shape[0]
print(f"Number of students from the Electronics Department: {electronics_students_count}")
```

```python
students_above_80_math = df[df['math_score'] > 80].shape[0]
print(f"Number of students who scored above 80 in math: {students_above_80_math}")
```
